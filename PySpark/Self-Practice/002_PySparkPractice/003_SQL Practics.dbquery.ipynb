```sql
CREATE PROCEDURE workspace.human_resources.get_emp_department_details (
  empid INT
)
SQL SECURITY INVOKER
BEGIN

  SELECT 
    e.BusinessEntityID,
    e.LoginID, 
    e.JobTitle,
    e.Gender,
    e.HireDate,
    d.Name AS DepartmentName,
    d.DepartmentID,
    edh.ShiftID,
    edh.StartDate,
    edh.EndDate
  FROM 
    workspace.human_resources.employee e
  INNER JOIN 
    workspace.human_resources.employee_department_history edh ON e.BusinessEntityID = edh.BusinessEntityID
  INNER JOIN 
    workspace.human_resources.department d ON edh.DepartmentID = d.DepartmentID
  WHERE 
    e.BusinessEntityID = empid;

END;
```
