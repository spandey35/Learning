```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Data Structure: [restaurant_id , 2021, 2022, 2023, 2024]
sales_data=  np.array([
    [1, 150000, 180000, 220000, 250000],  # Briyani
    [2, 120000, 140000, 160000, 190000],  # bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger 
    [5, 160000, 185000, 205000, 230000],  # Chai
    ])

print("== Sales Analysis==")
print("Shape of sales_data", np.shape(sales_data))
print("\n Sample data for 1st 3 restaurant are: ", sales_data [:3])
print("\n Sample data range for restaurant are: ", sales_data [:,1:])
```

```python
## Total Sales per Year
print(np.sum(sales_data, axis=0))
yearly_total = (np.sum(sales_data[:, 1:],axis=0))
print(yearly_total)
print(sum(yearly_total))


# Miniumn Sales per Year
Min_sales = (np.min(sales_data[:,1:], axis=1))
print("Min Sales", Min_sales)


# Maximun Sales per Year
Max_Sales_Year= (np.max(sales_data [: ,1:],axis=0))


# Avarage Sales per Restaurant
avg_sale = (np.mean(sales_data [:,1:], axis=1))
print("Ava_Sales", avg_sale)


# Cummulative Sales
cumm_sales = np.cumsum(sales_data [:,1:],axis=1)
print("Cumm_Sales", cumm_sales)

```

## Matplotlib Some Exmaple

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumm_sales,axis=0))
plt.title ("Avarage Cumulative Sales per year for Restaurants")
plt.ylabel("sales")
plt.xlabel("Years")
plt.grid(True)
plt.show()
```
