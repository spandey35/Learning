```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Suraj Pandey\OneDrive\Desktop\Learning\Python\Pandas\Excel Files\pokemon_data.csv')
df["Total"] =df.iloc[:,4:10].sum(axis=1)

cols = list(df.columns.values)
df = df [cols[0:4] + [cols[-1]] + cols [4:12]]

df.head()
```

```python
import re

regex=df.loc[df["Type 1"].str.contains("Fire|Grass" ,regex=True )]
regex
```

```python
## Let say we need to ignore the case. 

ignore_case=df.loc[df["Type 1"].str.contains("Fire|grass",flags=re.I ,regex=True )]
ignore_case
```

```python
## Let says if we wanted to give any spaces between 'Fire | grass'

import re

pattern = 'Fire | grass'
terms = [t.strip() for t in pattern.split('|')]
clean_pattern = '|'.join(terms)

ignore_case = df.loc[df["Type 1"].str.contains(clean_pattern, flags=re.I, regex=True)]
ignore_case
```
