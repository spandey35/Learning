```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Suraj Pandey\OneDrive\Desktop\Learning\Python\Pandas\Excel Files\pokemon_data.csv')
df["Total"] =df.iloc[:,4:10].sum(axis=1)

cols = list(df.columns.values)
df = df [cols[0:4] + [cols[-1]] + cols [4:12]]

df.head()
```

```python
## filtering the dataset

new_df=df.loc[((df["Type 1"]=="Grass") | (df["Type 2"]=="Poison")) & (df["HP"]>110)]
new_df.reset_index()
new_df
```

```python
## after filter , if we wanted to remove the index so we can do as below

new_df.reset_index(drop=True , inplace=True)
new_df
```

```python
## filter data fame which having name as mega 

WithMega=df.loc[(df["Name"].str.contains("Mega"))]
WithMega.head(5)
```
